# **Supermarket dataset analitikasi**





```python
# kutubxonalar
import pandas as pd
import plotly.express as px
```

```python
# faylni tortish va DataFrame ga aylantirish
url = "/content/drive/MyDrive/Data analyst learning/real_projects/karzinka_sovdo_analitic/sotuvlar.csv"
df = pd.read_csv(url)
del df['Invoice ID']
del df["Time"]
df.set_index("Date", inplace=True)
df.index = pd.to_datetime(df.index)

df
```

```python
# Datani filyallarga ajratish
a, b, c = df[df.Branch == "A"], df[df.Branch == "B"], df[df.Branch == "C"]
a = a.reset_index().groupby("Date").sum()
b = b.reset_index().groupby("Date").sum()
c = c.reset_index().groupby("Date").sum()
```

# **Asosiy qism**

```python
# Dataga Month column yaratish qismi
oy = df.reset_index().groupby("Date").sum()
yan, feb, mart = oy.iloc[:31], oy.iloc[31:59], oy.iloc[59:]
yan["Month"], feb["Month"], mart["Month"] = "January", "February", "March"
re_df = pd.concat([yan, feb, mart])
```

```python
# Dataga Week column yaratish qismi
weeks = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
re_df["Week"] = "week"
n = 0
for i in range(len(re_df.index)):
    if i == 0:
        re_df["Week"][i] = weeks[1]
        n += 1
    elif n == 7:
        n = 0
        re_df["Week"][i] = weeks[n]
    else:
        re_df["Week"][i] = weeks[n]
    n += 1

```

```python
# Shaharlar Yangon, Naypyitaw	va Mandalay	
fig1 = px.pie(df, values="gross income", names="Branch", width=1000, height=500, title="Bu filyallarning 3 oydagi sof daromadining (%)foizdlar")
fig1.show() 
```

```python
fig1 = px.pie(re_df, values="gross income", names="Month", width=1000, height=500, title=("Bu filyallarning 2019 yil Yanvar, Febral va Mart oylardagi sof daromad (%)foizlar "))
fig1.show()
```

```python
fig = px.bar(re_df, y="gross income", color="Month",  width=1000, height=600, title="Bu 3 oyning har bir kunidagi daromad statistikasi.")
fig.show()
```

```python
we = re_df.groupby("Week").sum().reset_index()
fig = px.bar(we,  y="Week", x="Total", orientation='h', color='Week' , width=1000, height=600, title="Bu 3 oydagi haftalik filyallarning umumiy pul aylanmasini statistikasi")
fig.show()
```

```python
fig = px.histogram(df, x="Payment", color="Branch", barmode="group", width=1000, height=600, title="Bu filyallaring to'lov turlari statistikasi")
fig.show()
```
